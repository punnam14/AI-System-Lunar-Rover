### Q6 (8 points)
Write a program which will find all such numbers which are divisible by 7 but are not a multiple of 5,
between 2000 and 3200 (both included). 
The numbers obtained should be printed in a comma-separated sequence on a single line.

````python
#initialize a list
result = []

#loop for iterating through the range 
for n in range(2000,3201):
    if n % 7 == 0 and n % 5 != 0:
        result.append(str(n))

#print the result by joining into a string using the comma as a separator
print(','.join(result))
````

### Q10 (5 points)
Write a program to print the following using while loop

> a. First 10 Even numbers

> b. First 10 Odd numbers


````python
#Prints the first 10 odd and even numbers
even = []
odd = []

num = 1

while num <= 10:
    even.append(2*num)
    odd.append(2*num - 1)
    
    num+=1
    
print('First 10 Even Numbers are :',even)
print('First 10 Odd Numbers are :',odd)
````
